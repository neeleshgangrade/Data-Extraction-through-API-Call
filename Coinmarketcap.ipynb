```python
#importing required libraries
import requests
import pandas as pd
import json
from prettytable import PrettyTable
```

```python
#Creating prettytable
tableobj = PrettyTable()
key_value = 'eb3c61fc-f474-40e7-af52-314c8e7e13d7'
```

```python
#defining the key value to access the data from coinmarket
api_endpoint = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest?CMC_PRO_API_KEY='
api_endpoint += key_value
api_endpoint
```

```python
#getting the data in json file
requests.get(api_endpoint).json()
```

```python
#Storing the json data in a new variable
json_data_storing = requests.get(api_endpoint).json()
json_data_storing
```

```python
#Getting data for the specific 'data' key
CryptoData = json_data_storing['data']
```

```python
for currency in CryptoData:
    currency_name = currency['name']
    currency_price = currency['quote']['USD']['price']
    percent_change_1hour = currency['quote']['USD']['percent_change_1h']
    percent_change_24hour = currency['quote']['USD']['percent_change_24h']
    percent_change_7days = currency['quote']['USD']['percent_change_7d']
    percent_change_30days = currency['quote']['USD']['percent_change_30d']
    tableobj.add_row([currency_name, currency_price, percent_change_1hour, percent_change_24hour, percent_change_7days, percent_change_30days])
    
```

```python
#Renaming the columns
tableobj.field_names = ["Currency Names", "Currency Price", "Currency Percentage Change 1_hour", "Currency Percentage Change 24_hour",
                       "Currency Percentage Change 7_days", "Currency Percentage Change 30_days"]
```

```python
#Having look at the generated table
tableobj
```

```python
#writing the extracted table to text file
table_txt = tableobj.get_string()
with open('output.txt', 'w') as file:
    file.write(table_txt)
```
